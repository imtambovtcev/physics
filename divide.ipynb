```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: x**3+3*x+3
```

```python
i=7
a=-1
b=0
x=[-1,0.]
for _ in range(i):
    x.append(0.5*(a+b))
    if f(x[-1])>0:
        b=x[-1]
    else:
        a=x[-1]
x=np.array(x)
print(f'{x = }')
    
```

```python
X=np.linspace(-1.,0.,100)
plt.plot(X,f(X))
plt.grid()
plt.savefig('empty.png', transparent=True)
plt.show()
plt.plot(X,f(X))
plt.plot(x[-1],f(x[-1]),'rX')
plt.grid()
plt.savefig('empty_a.png', transparent=True)
plt.show()
plt.plot(X,f(X),'bX')
plt.grid()
plt.plot(x[-1],f(x[-1]),'rX')
plt.savefig('empty_x.png', transparent=True)
plt.show()
```

```python
plt.plot(X,f(X))
plt.plot(x,f(x),'rX')
plt.grid()
plt.savefig('dots{}.png'.format(i), transparent=True)
plt.show()
```
